## Array 
- Linear data structure

- continuous in memory

- Elements accessed randomly by use of indexing

- Homogeneous Elements (Similar elements)

### Application
- Store data in a linear fashion

- Suitable for applications that require frequent searching

### Types of Arrays

- One-dimensional arrays

- Two-dimensional arrays

- Multi-dimensional arrays

- Difficulty to inset and delete an element

#### One-dimensional arrays

- Single row or single column

- Elements are stored in a single row or column

        - Example: arr = [1, 2, 3, 4, 5]

- Access elements by index: arr[i]


#### Two-dimensional arrays

- Two rows and two columns

- Elements are stored in two rows and two columns

        - Example: [[1, 2], [3, 4]]

- Access elements by index: arr[i][j]

#### Multi-dimensional arrays

- More than two rows and columns

- Elements are stored in more than two rows and columns

        - Example: [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

- Access elements by index: arr[i][j][k]

### Advantages of arrays

- Easy to implement

- Easy to access elements

- Easy to manipulate elements

- Easy to search elements

### Disadvantages of arrays

- Fixed size

- Wastage of memory

- Inefficient memory usage

- Inefficient memory allocation


## Complexity of an array

- `Time complexity: O(1)`

- `Space complexity: O(n)`
